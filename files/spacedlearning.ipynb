## Background and Motivation

The motivation behind our project was a shared interest in learning and memory science. Last summer Elliott worked at a Cambridge based startup called <a href='http://www.firecracker.me'> Firecracker</a>, an online learning platform for medical students. Firecracker helps its users prepare for medical exams, including the United States Medical Licensing Exam, by giving them access to content on each of the topics in these exams, quizzing them after they review material for the first time, and using a user's scores to create a customized review schedule. 

The quiz questions follow a flash card system. Most quiz questions are 'recall' questions where the user self reports their mastery of that question on an integer scale from 1 to 5. A score of 1 denotes not remembering the answer to a question at all and a score of 5 denotes knowing an answer perfectly. As users review topics in Firecracker and flag them for quizzing, they receive new quiz questions relating to the material they reviewed. A sample recall flash card question is included below.

<br><img src="files/recall_question.JPG"/><br><br>

Firecracker uses a user's review scores to create a customized review schedule. The review schedule is created using a spaced learning algorithm in order to optimize learning. For example, if a user scores a 1 on a question they are scheduled to see the question sooner than if they had scored a 5. The concept of spaced repetition is a learning technique that incorporates increasing intervals of time between subsequent reviews of previously learned material in order to exploit the psychological <a href='http://en.wikipedia.org/wiki/Spacing_effect'>spacing effect</a>. Our group was especially interested in exploring data for the application of this learning technique in a real world environment.

The following cell is to create columns that tell us how many times a user has seen a question before, the time elapsed since they last saw that question, and their previous score. If the user did not see the question before then their previous score is -1. Since this block of code takes some time to run, we ran it separately and saved it as a csv, which we access in the cell after this.

```python
#this cell is to create columns that tell us how many times a user has seen a question before, the time
#elapsed since they last saw that question, and their previous score. If the user did not see the question before
#then their previous score is -1

#sort by time created at so we can access the first time a suser answered a question
recalls_full = recalls_full.sort('created_at')

#create a new column for number of times seen
recalls_full['times_seen'] = [0] * len(recalls_full)

#create a new column for the time since last seen
recalls_full['time_elapsed'] = [0.] * len(recalls_full)

#create a new column for previous score
recalls_full['prev_score'] = [0] * len(recalls_full)

#group by user_id and question_id
grouped_uq = recalls_full.groupby(['user_id', 'question_id'])

for k, v in grouped_uq:
    #sort by time within each groupby
    v = v.sort('created_at')
    #keep track of how many times they have seen the question
    count = 0
    
    #iterate through all the times they have seen the question and update the times_seen column
    #additionally create a time_elapsed column since last seen
    for i, r in v.iterrows():
        recalls_full.times_seen.ix[i] = count
        if len(v) == 1:
            continue
        elif count == 0:
            recalls_full.time_elapsed.ix[i] = float('nan')
            recalls_full.prev_score.ix[i] = -1
        else:
            end = np.datetime64(r['created_at'])
            begin = np.datetime64(v.loc[prev_index]['created_at'])
            recalls_full.time_elapsed.ix[i] = float((end - begin)/np.timedelta64(1, 'D'))
            recalls_full.prev_score.ix[i] = v.loc[prev_index]['recall_score']
        count += 1
        prev_index = i
        
recalls_full.head()
```

## Final Analysis

Our primary objective was to find a predictive model for the recall scores of a user, given information about their history. This predictive model allows us to test how well the Ebbinghaus forgetting curve is represented in our data, to see if our users trends actually follow this model. Furthermore, a predictive model allows Firecracker to more effectively refine its spaced learning model, as it helps the system determine when a user has dropped to a certain threshhold of understanding, at which point it is most effective to reintroduce review material. Even more precisely, if we are able to obtain a good model for not only the decay of memory over time, but also the benefit of revising material, this information can be used to optimize the spaced learning algorithm Firecracker uses.

### An Initial Model

Consider an initial model, which takes

$$\hat Y_{predicted} = \hat \mu + \hat \theta_u + \hat \gamma_q$$

where the unknown parameters $\theta_u$ and $\gamma_q$ indicate the deviations/biases of user $u$ and question $q$, respectively, from some intercept parameter $\mu$.

These unknown parameters need to be fit. We'll replace them with their "mean" estimates from the data.

$$\hat Y_{predicted} = \bar Y + (\bar Y_u - \bar Y) + (\bar Y_q - \bar Y)$$

where $\bar Y_u$ is the user's average ratings and $\bar Y_q$ is the question's average ratings, and $\bar Y$ is the average rating over all reviews. Our model runs this iteratively throughout the entire course of our data (sorted by time), and computes results based on all the previous history that we can see at that point in time. For the first prediction, before we have any data, we'll use a predicted value of 3. For predictions where we don't have any previous user data or previous question data, we'll assume the difference between the average user ratings and the average overall ratings is 0, or that the difference between the average question ratings and the average overall ratings is 0.

This model attempts to account for biases from specific users and questions. We hope to use it as a baseline model that we can compare our other models to. The following function builds a dictionary of predictions according to what is described above.

```python
"""
Implement an initial model that performs as described above.

Parameters:
    recalls: the recalls dataframe to use

Returns a dictionary of predictions where the key is the user_id and the value
is an array of predictions for the user's scores for all the questions that user
has answered, sorted by the time that the user answered each question.
"""
def initial_model(recalls):
    # sort by time
    recalls = recalls.sort(columns = 'created_at')
    
    # keep track of sum of scores and the number of scores so we can calculate overall average
    total = 0
    count = 0
    # the predictions dictionary
    predictions = {}
    # dictionary of actual scores for users, where each key is user_id and each value is an ordered tuple
    # of total score and number of questions answered
    actual = {}
    # dictionary of actual scores for questions, where each key is question_id and each value is an ordered
    # tuple of total score and number of times that question has been answered
    questions = {}
    for k, recall in recalls.iterrows():
        # if we just started, begin with a prediction of 3
        if len(predictions) == 0:
            predictions[recall["user_id"]] = [3.]
            actual[recall["user_id"]] = (recall["recall_score"], 1)
            questions[recall["question_id"]] = (recall["recall_score"], 1)
        # otherwise, calculate predictions based on entire previous history
        else:
            avg = float(total) / float(count)
            
            # add keys and values into dictionaries if the key is not present
            if recall["user_id"] not in predictions:
                predictions[recall["user_id"]] = []
            if recall["user_id"] not in actual:
                actual[recall["user_id"]] = (0, 0)
            if recall["question_id"] not in questions:
                questions[recall["question_id"]] = (0, 0)
            
            if actual[recall["user_id"]][1] == 0:
                du = 0.
            else:
                du = (float(actual[recall["user_id"]][0]) / float(actual[recall["user_id"]][1])) - avg
            
            if questions[recall["question_id"]][1] == 0:
                dq = 0.
            else:
                dq = (float(questions[recall["question_id"]][0]) / float(questions[recall["question_id"]][1])) - avg
            
            prediction = int(round(avg + du + dq))
            if prediction > 5:
                prediction = 5
            if prediction < 1:
                prediction = 1
            
            # update dictionaries
            predictions[recall["user_id"]].append(prediction)
            actual[recall["user_id"]] = (actual[recall["user_id"]][0] + recall["recall_score"], actual[recall["user_id"]][1] + 1)
            questions[recall["question_id"]] = (questions[recall["question_id"]][0] + recall["recall_score"], questions[recall["question_id"]][1] + 1)
        total += recall["recall_score"]
        count += 1
    return predictions
```

The following function calculates error for our predictions by comparing them to the actual values in the recalls dataframe. The error is the sum of the squared differences between predictions and actual values.

```python
"""
Calculates error for predictions by comparing them to the actual values in the recalls dataframe.

Parameters:
    predictions: the predictions dictionary, obtained from the previous function
    recalls: the recalls dataframe to use
    burn_in (optional): the fraction of predictions to discard per user; discarded predictions will
        come from the beginning of the list of predictions; default 0

Returns an error value calculated in the way described above.
"""
def calc_error(predictions, recalls, burn_in=0):
    # sort by time
    recalls = recalls.sort(columns = 'created_at')
    
    error = 0
    for user in predictions:
        actual = recalls[recalls.user_id == user].recall_score.values
        actual = actual[int(round(len(actual) * burn_in)):]
        for i in range(0, len(actual)):
            error += (actual[i] - predictions[user][i])**2
    return error
```

### An Ebbinghaus Model

We consider instead a different model, based off of the Ebbinghaus forgetting curve. We discover that there are two factors in our model: a coefficient $C$ which represents the benefit of seeing a memory ($C$ is, intuitively, the precent of knowledge retained upon reviewing material), and a parameter $W$ which represents the rate at which the mastery is lost, so that the overall modified Ebbinghaus forgetting curve is given by
$$R(t) = Ce^{-Wt}$$

Being able to predict $C$ and $W$ well for a given rating gives us insight into how to best help the user to learn using Firecracker.

First we filter out recall scores whose time_elapsed is very high, i.e. scores where it has been a long time since the user last saw the question. We do this because when time_elapsed is high, it becomes increasingly unlikely that the user actually has not reviewed the material (outside of Firecracker) within that time span, and furthermore when we fit a line which is forced through the origin, such values are weighted too highly.

```python
recalls = pd.read_csv('recalls_new.csv')
#we try to filter out questions whose time_elapsed is very high
print(len(recalls))
recalls = recalls[recalls.time_elapsed < 25]
print(len(recalls))
```

####Factors in our model####

The standard model for memory retention is given by the Ebbinghaus forgetting curve, an exponential curve $e^{-\frac tS}$ where $S$ is a parameter which describes the relative strength of the memory. Actually, for ease of notation, we'll use the equivalent model $e^{-tW}$ where $W$ is simply $1/S$, a parameter which describes the relative weakness of the memory. We are interested in understanding which parameters determine $S$, which are the parameters on which we will condition. The four factors we have been considering are as follows:

- User

    - it is possible that some users are simply better at retaining knowledge than others

- Question

    - it is possible that some questions are simply harder to remember than others

- Number of times seen

    - we would hope that as the user sees the same question more times, the duration of the user's retention of the material is improved

- Previous rating

    - the above plots indicate that mastery is correlated with persistence: even though the second set of plots indicates that higher-rated questions are repeated after a much greater timespan has elapsed, as one would expect of spaced learning algorithms, the first set of plots indicates that mastery persists much longer, as even when reviewed much later, high initial ratings are still strongly correlated with high ratings upon repetition.

It makes sense that the latter two would be positively correlated with one another, since seeing a question more times should lead to a higher rating, and vice versa, higher ratings should indicate more experience with the material. We saw in the exploratory analysis that these are indeed highly correlated, as the user-reported ratings improve with the number of times they have seen a given question. We'll try each in turn.

####Training and test data####

Because our data is temporal, there is a natural way to consider 'training' and 'test' data sets, namely that in making each prediction, we can consider the previous ratings to be given. The chunks of data we are interested in split naturally into sequences of ratings in which the user and question are the same. We will assume certain global information about a user's overall retention and a question's difficulty are known (i.e. we will calculate it once for the entire data set and take those as given, in order to reduce the amount of computation required).

Our baseline model will look something like this: if $W$ is a baseline, global weakness parameter, then our prediction will be

$$e^{-tW} \frac{e^{-tW_q}}{e^{-tW}} \frac{e^{-tW_u}}{e^{-tW}} \frac{e^{-t W_h}}{e^{-tW}} = e^{-(W_1 + W_u + W_h - 2W)t}$$

where $W_q$ is a measure of the difficulty of the question (higher $W_q$ indicates that a question is harder, and more easily forgotten), $W_u$ is a measure of the weakness of the user's memory (higher $W_u$ indicares that a user has more trouble remembering information), and $W_h$ is some correction which is a function of a user's history with a specific question.

We anticipate that $W_q, W_u$ don't change much once there is some amount of history for a user, since we are only really interested in long term trends, and so we will compute $W_q, W_u$ once over the entire data set, without altering them prediction-by-prediction. $W_h$, on the other hand, is a parameter which changes with a given user's history with a particular question, which will need to be updated between predictions for a single (user, question) sequence of ratings - this is the parameter which will depend on the 'test' or historical data.

####Finding $W, W_u, W_q$####

To find $W$, we simply fit a basic forgetting curve (exponential decay) across all data, i.e. across all points (rating, time since last seen). We'll then find the 'correction factors' due to each particular user, the $W_u$, by fitting a forgetting curve across all ratings for a given user. We'll similarly find $W_q$ by fitting a curve across all ratings for a given question.

Denoting ratings $R$, and time $t$, we want to fit an exponent of the form $R = e^{-tW} \Rightarrow \log R = -tW$. Thus we perform a linear fit for $t, \log R$, the slope of which is $-W$.

```python
recalls = recalls.sort('created_at')
```

```python
def expo_fit(df):
    
    # find all (time since last review, rating) pairs
    t = []
    r = []
    
    for key, row in df.iterrows():
        if row['times_seen'] > 0:
            t.append(row['time_elapsed'])
            r.append(row['recall_score'])
    t = np.array(t)
    logr = np.log(np.array(r)/5.)
    
    # fit a line through the origin, using the formula sum(x_i*y_i)/sum(x_i**2)
    # citation http://web.ist.utl.pt/~mcasquilho/compute/errtheory/,regression/regrthroughorigin.pdf
    return -np.dot(logr, t)/np.dot(t, t)
```

```python
# we split the following few cells because they are computationally intensive, so that we can run them separately
# hard code W so we don't have to run this again, toggle commmenting as needed
W = expo_fit(recalls)
print(W)
```

```python
def fit_users(df):
    user_dict = {}
    grouped = df.groupby('user_id')
    for user, user_df in grouped:
        user_dict[user] = expo_fit(user_df)
    return user_dict
        
def fit_questions(df):
    q_dict = {}
    grouped = df.groupby('question_id')
    for q, q_df in grouped:
        q_dict[q] = expo_fit(q_df)
    return q_dict

W_u = fit_users(recalls)
W_q = fit_questions(recalls)
```

####Modifying our model####

Let's try instead a different model, where we eliminate the $e^{-Wt}$ dependence in our decay curves by instead taking just a geometric mean, so that we have, considering only the user and question,
$$R = e^{-t(W_u + W_q)/2}$$
and if we consider the additional question history,
$$R = e^{-t(W_u + W_q + W_h)/3}$$

```python
def geom_predict(df, W_u, W_q, recalls_mean, u_means, q_means):
    #prediction is a dictionary in which keys are the dataframe keys, and values are the predicted ratings
    prediction = {}
    grouped_uq = df.groupby(['user_id', 'question_id'])

    for k, v in grouped_uq:
        W_user = W_u[k[0]]
        W_question = W_q[k[1]]
        for i, j in v.iterrows():
            if j['times_seen'] == 0:
                prediction[i] = min(max(u_means[k[0]] + q_means[k[1]] - recalls_mean, 1), 5)
            else:
                prediction[i] = np.exp(-j['time_elapsed']*(W_user + W_question)/2)*5
                if prediction[i] > 5:
                    print(prediction[i], W_user, W_question, W)
                    return 1
    
    return prediction
```

```python
simple_geom_predictions = geom_predict(recalls, W_u, W_q, recalls_mean, u_means, q_means)
```

The error on this model is not significantly improved, but we still prefer this version because it does not predict values above 5. As mentioned before, we have found that $W$ is not actually a meaningful characterization of the entire data set, and so we choose to avoid dependence of our model on $W$.

There remains a problem in which the predictions are very heavily skewed to the right, rather than resembling the actual distribution of the data, as can be seen from the histograms above. This is in part due to the fact that again, points far from the origin, e.g. points with high ratings, tend to be weighted very heavily, since we have forced our (log-)linear fit through the origin; this means that points with very high ratings, especially those with very large values for time_elapsed, make the decay factor much lower than it should be (since those points which indicate the least decay are weighted most heavily).

In order to adjust for this, we try instead the same method for calculating $W_u, W_q$, but we actually allow the linear regression to choose any $y$ intercept it pleases.

```python
def expo_fit_new(df):

    # find all (time since last review, rating) pairs
    t = []
    r = []
    
    for key, row in df.iterrows():
        if row['times_seen'] > 0:
            t.append(row['time_elapsed'])
            r.append(row['recall_score'])
    t = np.array(t)
    logr = np.log(np.array(r)/5.)
    
    if len(t) == 0:
        return None
    # fit a line through the origin, using the formula sum(x_i*y_i)/sum(x_i**2)
    # citation http://web.ist.utl.pt/~mcasquilho/compute/errtheory/,regression/regrthroughorigin.pdf
    return list(np.polyfit(t, logr, 1))

def fit_users_new(df):
    user_dict = {}
    grouped = df.groupby('user_id')
    for user, user_df in grouped:
        user_dict[user] = expo_fit_new(user_df)
    return user_dict
        
def fit_questions_new(df):
    q_dict = {}
    grouped = df.groupby('question_id')
    for q, q_df in grouped:
        q_dict[q] = expo_fit_new(q_df)
    return q_dict

W_u = fit_users_new(recalls)
W_q = fit_questions_new(recalls)
```

We can think of this coefficient, which is $R(t = 0)$, the rating predicted if the user is immediately asked the same question again, as a baseline retention factor for the information: even right after reading a piece of sufficiently complex information, a student might not remember it perfectly.

We have two ways to make predictions:
$$R_u = C_u e^{-tW_u}, R_q = C_q e^{-tW_q}$$
where $W_u, W_q$ are the decay coefficients for the given user and question, and $C_u, C_q$ are parameters which describe the ability of a user to absorb the material upon reviewing it and the corresponding difficulty of the material referred to by a question, respectively. Essentially, the W's correspond to how well information is retained, while the C's describe the amount of mastery gained by reviewing the material.

```python
def unlock_intercept_predict(df, W_u, W_q, recalls_mean, u_means, q_means):
    #prediction is a dictionary in which keys are the dataframe keys, and values are the predicted ratings
    prediction = {}
    grouped_uq = df.groupby(['user_id', 'question_id'])

    for k, v in grouped_uq:
        W_user = W_u[k[0]]
        W_question = W_q[k[1]]
        for i, j in v.iterrows():
            if j['times_seen'] == 0:
                prediction[i] = min(max(u_means[k[0]] + q_means[k[1]] - recalls_mean, 1), 5)
            else:
                prediction[i] = (W_user[1]*W_question[1])**.5 *np.exp(j['time_elapsed']*(W_user[0] + W_question[0])/2)*5
    
    return prediction
```

```python
unlocked_intercept_predictions = unlock_intercept_predict(recalls, W_u, W_q, recalls_mean, u_means, q_means)
```

Although the spread in our predictions is still not as large as we would like, we have indeed eliminated the problem in which the predictions are skewed too far to the right. This is strong evidence that the problem with our former model is indeed, as we hypothesized, due to the disproportionate weight on the data points far from the origin, which resulted from forcing our log-linear fit through the origin. Weakening this restriction not only allows us to weigh our data more evenly, but also enables us to consider a more complex model for memory. The Ebbinghaus forgetting curve assumes that when material is reviewed, retention at that moment is 100%, but he was working with simple concepts like memorizing strings of digits, for which that assumption is reasonable. For us, we are dealing with much more complex concepts, in which, as previously discussed, reading the answer to a problem does not immediately lead to mastery of the material, even in the short term.

###Weighting $W_h$
We've been assuming that we want to weight all of these three factors $W_u, W_q, W_h$ evenly. It's reasonable to assume that question and user values have more or less the same predictive power, but history may or may not be so. We'll next try the model

$$R(t) = (C_u*C_q*C_h^a)^{1/3} e^{-t(W_u + W_q + a*W_h)/3}$$

where $a$ is some weight which we give to the $W_h$.

Since we're now fitting this parameter $a$, we want to be sure we're not overfitting, so at this point it becomes necessary to split our data in to test and training sets.

```python
import random
def split_df(df):
    test = []
    train = []
    
    #fitting 'a' will take a very long time, so we'll try to do it on a small subset of the data first, 10% of the total data
    #we put each user/question pair in the test or training set with 10% probability
    for k, v in df.groupby(['user_id', 'question_id']):
        r = random.random()
        if r < 0.1:
            test.append(v)
        elif r > 0.9:
            train.append(v)
    
    return pd.concat(test), pd.concat(train)

test, train = split_df(recalls)
```

###Remaining problems
We have achieved fairy substantial success in applying the Ebbinghaus forgetting curve model to predict recall scores. The remaining discrepancies are primarily due to the simple fact that there are too many factors involved to model correctly.

Firecracker is not a controlled experiment. The spaced learning algorithm of Firecracker itself interferes with our ability to fit curves: high scores are not rescheduled for 10-20 days, which means our data is not distributed evenly across the time span. Furthermore, again for these high values, it is likely that students have reviewed the material outside of Firecracker during this time, and so the 5's are artificially inflated by that as well, since students forget less than the exponential forgetting curve would expect them to. This in part explains the underrepresentation of scores of 5 in our model.

Most mid-low ratings are rescheduled for the near future, in the next few days. This means that points near $t = 0$ are distributed fairly evenly between 1, 2, 3, and the fit exponential will have time to decay only to 2 or 3 in this short period of time. This explains the underrepresentation of scores of 1, which is also in part an artifact of the spaced learning schedule that Firecracker uses.

Also note that our model is not probabilistic in nature: our model simply predicts the mean of what should be a distribution of possible values. This also prohibits the model from predicting extreme values, due to regression effects.

### A Latent Factor Model For Predicting User Scores

Before concluding, we wish to check one final model to see if we can do better than the previous two. Previously, we have predicted user recall scores by implementing procedures which involved a number of somewhat arbitrary choices. We also wish to explore what happens if we start with a probabilistic model for how the data were generated. This takes an approach that is very similar to that used in Question 3 of Homework 4. We construct a model of how recall scores are generated, and use this model to build a prediction system. As in Homework 4, we compute the posterior distributions of the model's parameters using a Gibbs sampler. We define terms in the same way as in Homework 4, with the exception that the items are now questions instead of restaurants. Note that the way in which we draw for $\mu$, $\sigma$, $\gamma_m$, and $\theta_u$ are the same as in Homework 4. Also note that for simplicity, we will limit our data to only those rows where it is the second time that a user has seen the given question. This will allow us to run one latent factor model and see how we do.

The draws and sampler code are implemented below.

```python
"""
Function
--------
gamma_m_draw

Draw a single sample from the conditional posterior distribution
of gamma_m.

Inputs
-------
X_m: A g-by-L+1 matrix, defined in Homework 4. 
Y_m: A 1D vector of length g, defined in Homework 4.
sig2: Residual _variance_, as defined in Homework 4.
Lambda_gamma: Prior precision matrix.

Outputs
--------
Single draw from conditional posterior, defined in Homework 4.
"""
#Item-specific parameters given all else
def gamma_m_draw(X_m, Y_m, sig2, Lambda_gamma):
    # calculate Q_m, as defined above
    Q_m = (X_m.transpose().dot(X_m) / sig2) + Lambda_gamma
    
    # return the correct MVN
    return np.random.multivariate_normal(LA.inv(Q_m).dot(X_m.transpose()).dot(Y_m) / sig2, LA.inv(Q_m))
```

```python
"""
Function
--------
theta_u_draw

Draw a single sample from the conditional posterior distribution
of theta_u.

Inputs
-------
X_u: A g-by-L+1 matrix, defined in Homework 4. 
Y_u: A 1D vector of length g, defined in Homework 4.
sig2: Residual _variance_, as defined in Homework 4.
Lambda_theta: Prior precision matrix.

Outputs
--------
Single draw from conditional posterior, defined in Homework 4.
"""
#User-specific parameters given all else
def theta_u_draw(X_u, Y_u, sig2, Lambda_theta):
    # calculate Q_u, as defined above
    Q_u = (X_u.transpose().dot(X_u) / sig2) + Lambda_theta
    
    # return the correct MVN
    return np.random.multivariate_normal(LA.inv(Q_u).dot(X_u.transpose()).dot(Y_u) / sig2, LA.inv(Q_u))
```

```python
"""
Function
--------
factor_gibbs

Runs a gibbs sampler to infer mean, variance, user-specific, and item-specific
parameters.

Inputs
-------
data: A dataframe containing ratings data.
L: Dimension of latent factors.
maxit: Number of samples to draw from posterior.
Lambda_theta_diag: Hyperparameter controlling regularization of Theta.
Lambda_gamma_diag: Hyperparameter controlling regularization of Gamma.
progress: if true, print iteration number every 100 iterations.

Outputs
--------
Dictionary with elements
mu: Draws of mu. 1D array of length maxiter.
sig2: Draws of sig2, residual _variance_. 1D array of length maxiter.
theta: Draws of Theta. U-by-L-by-maxiter array.
gamma: Draws of Gamma. M-by-L-by-maxiter array.
EY: Draws of fitted values of Y. N-by-maxiter array.
"""
def factor_gibbs(data, L, maxit, Lambda_theta_diag, Lambda_gamma_diag, progress=True):
    data = data.copy()
    N = data.shape[0]

    #Create indices that allow us to map users and questions to rows
    #in parameter vectors.
    uusers, uidx = np.unique(data.user_id, return_inverse=True)
    uitems, midx = np.unique(data.question_id, return_inverse=True)

    nusers = uusers.size
    nitems = uitems.size

    #Add numerical indices to dataframe.
    data["uidx"] = uidx
    data["midx"] = midx

    #Group observations by user and by question.
    ugroups = data.groupby("uidx")
    mgroups = data.groupby("midx")

    all_avg = data.recall_score.mean()
    u_avg = ugroups.recall_score.mean()
    m_avg = mgroups.recall_score.mean()

    #Initialize parameters and set up data structures for
    #holding draws.
    #Overall mean
    mu = all_avg
    mu_draws = np.zeros(maxit)
    #Residual variance
    sig2 = 0.5
    sig2_draws = np.zeros(maxit)

    #Matrix of user-specific bias and L latent factors.
    theta = np.zeros([nusers, L+1])
    theta[:,0] = u_avg-all_avg
    theta_draws = np.zeros([nusers, L+1, maxit])

    #Matrix of item-specific bias and L latent factors.
    gamma = np.zeros([nitems, L+1])
    gamma[:,0] = m_avg-all_avg
    gamma_draws = np.zeros([nitems, L+1, maxit])

    #Matrix for holding the expected number of stars
    #for each observation at each draw from the posterior.
    EY_draws = np.zeros([data.shape[0], maxit])

    #Inverse covariance matrices from the prior on each theta_u
    #and gamma_b. These are diagonal, like Ridge regression.
    Lambda_theta = np.eye(L+1)*Lambda_theta_diag
    Lambda_gamma = np.eye(L+1)*Lambda_gamma_diag

    #Main sampler code
    for i in range(maxit):
        if i%100==0 and progress:
            print i

        #The entire regression equation except for the overall mean.
        nomu = np.sum(theta[data.uidx,1:]*gamma[data.midx,1:], axis=1) +\
                  theta[data.uidx,0] + gamma[data.midx,0]

        #Compute the expectation of each observation given the current
        #parameter values.
        EY_draws[:,i]=mu+nomu

        #Draw overall mean from a normal distribution
        mu = np.random.normal(np.mean(data.recall_score-nomu), np.sqrt(sig2/N))
        #Draw overall residual variance from a scaled inverse-Chi squared distribution.
        sig2 = np.sum(np.power(data.recall_score-nomu-mu,2))/np.random.chisquare(N-2)
        
        #For each item
        for mi,itemdf in mgroups:
            #Gather relevant observations, and subtract out overall mean and
            #user-specific biases, which we are holding fixed.
            Y_m = itemdf.recall_score-mu-theta[itemdf.uidx,0]
            #Build the regression design matrix implied by holding user factors
            #fixed.
            X_m = np.hstack((np.ones([itemdf.shape[0],1]),
                             theta[itemdf.uidx,1:]))
            gamma[mi,:] = gamma_m_draw(X_m, Y_m, sig2, Lambda_gamma)
            
        #For each user
        for ui,userdf in ugroups:
            #Gather relevant observations, and subtract out overall mean and
            #question-specific biases, which we are holding fixed.
            Y_u = userdf.recall_score-mu-gamma[userdf.midx,0]
            #Build the regression design matrix implied by holding question factors
            #fixed.
            X_u = np.hstack((np.ones([userdf.shape[0],1]),
                             gamma[userdf.midx,1:]))
            
            theta[ui,:] = theta_u_draw(X_u, Y_u, sig2, Lambda_theta)

        #Record draws
        mu_draws[i] = mu
        sig2_draws[i] = sig2
        theta_draws[:,:,i] = theta
        gamma_draws[:,:,i] = gamma

    return {"mu": mu_draws, "sig2": sig2_draws,
            "theta": theta_draws, "gamma": gamma_draws,
            "EY": EY_draws}
```

### Differences between Institutions

Another question that may be interesting is whether users from different institutions tend to perform differently. For example, do users from one institution do better initially than users from another institution? The answer to this question could give us insights on the strengths and weaknesses of institutions, and could also be used to help Firecracker suggest questions to users by using knowledge of the user's institution to target questions in which that user is likely to perform worse initially, so that users could improve more.

One way that we could answer this question is by examining the common support of questions that each user has answered, create a graph where each node is a user, and determine whether users from the same institution tend to clump together.

To make sure we have a large enough common support, we examine only those users who have answered at least a certain number of unique questions. We'll refer to these users as "major users" of the Firecracker platform.

We also need to ensure that the institutions we consider are well-represented, i.e. that we have at least a certain number of major users per institution. We screen out institutions where we have very few major users, and only consider institutions where there are at least a certain number of major users. We'll call the institutions that we include "major institutions" of the Firecracker platform.

The following function prints out some of the information that we are interested in.

```python
"""
Given restraints on how many unique questions users need to have answered to be considered a major user
and how many major users institutions need to have to be considered a major institution, print some useful
summary statistics. Also return the dataframe of all entries that involve questions in the common support.

Parameters:
    recalls: the recalls dataframe to use
    nq: number of unique questions users need to have answered to be considered a major user
    nu: number of major users institutions need to have to be considered a major institution

Returns recalls_new, a dataframe of all entries that involve questions in the common support.
"""
def filter_recalls(recalls, nq, nu):
    # filter recalls to only include users who have answered at least nq questions
    grouped_u = recalls.groupby('user_id')
    recalls_new = grouped_u.filter(lambda x: x.question_id.nunique() >= nq)
    
    print "Number of questions answered by major users: " + str(len(recalls_new))
    grouped_u = recalls_new.groupby('user_id')
    print "Number of major users: " + str(grouped_u.ngroups)
    
    # filter recalls_new to only include institutions where there are at least nu major users
    grouped_i = recalls_new.groupby('institution_id')
    recalls_new = grouped_i.filter(lambda x: x.user_id.nunique() >= nu)
    
    print "Number of questions answered by major users in major institutions: " + str(len(recalls_new))
    grouped_i = recalls_new.groupby('institution_id')
    print "Number of major institutions: " + str(grouped_i.ngroups)
    # number of users in major institutions
    n_users_final = len(set(recalls_new.user_id))
    print "Number of major users in major institutions: " + str(n_users_final)
    
    # find number of questions in the common support
    grouped_q = recalls_new.groupby('question_id')
    recalls_new = grouped_q.filter(lambda x: x.user_id.nunique() == n_users_final)
    print "Number of questions in common support: " + str(len(recalls_new))
    
    return recalls_new
```

There are two cases above where we have a nontrivial common support.

In the first case, the minimum number of subjects is 10 and the minimum number of major users is 5. The issue with this case is that there is only one major institution that satisfies these requirements, so we can't really get any sort of comparison between institutions.

In the second case, the minimum number of subjects is 15 and the minimum number of major users is 3. #### An Alternative Method

One possible alternative is the following:

In a manner similar to the first method we tried, only consider major users who have answered at least a certain number of unique questions, and major institutions who have at least a certain number of major users. Then for each pair of users in the remaining data set, consider the average difference between their scores on the questions that just those two users have in common. We take an average because the number of questions in common will likely vary across pairs. We can define a distance between two users as this average difference, and create a graph of users with this definition of distance.

The following function prints out some of the information we're interested in to help us determine good restraints to use.

```python
"""
Given restraints on how many unique questions users need to have answered to be considered a major user
and how many major users institutions need to have to be considered a major institution, print some useful
summary statistics. Also return the dataframe of all entries that remain after filtering by our requirements.

Parameters:
    recalls: the recalls dataframe to use
    nq: number of unique questions users need to have answered to be considered a major user
    nu: number of major users institutions need to have to be considered a major institution

Returns recalls_new, a dataframe of all entries that remain after filtering by our requirements.
"""
def filter_recalls4(recalls, nq, nu):
    # filter recalls to only include users who have answered at least nq questions
    grouped_u = recalls.groupby('user_id')
    recalls_new = grouped_u.filter(lambda x: x.question_id.nunique() >= nq)
    
    print "Number of questions answered by major users: " + str(len(recalls_new))
    grouped_u = recalls_new.groupby('user_id')
    print "Number of major users: " + str(grouped_u.ngroups)
    
    # filter recalls_new to only include institutions where there are at least nu major users
    grouped_i = recalls_new.groupby('institution_id')
    recalls_new = grouped_i.filter(lambda x: x.user_id.nunique() >= nu)
    
    print "Number of questions answered by major users in major institutions: " + str(len(recalls_new))
    grouped_i = recalls_new.groupby('institution_id')
    print "Number of major institutions: " + str(grouped_i.ngroups)
    # number of users in major institutions
    n_users_final = len(set(recalls_new.user_id))
    print "Number of major users in major institutions: " + str(n_users_final)
    
    # find minimum number of questions that any pair of users has in common
    min_num = len(recalls_new)
    # dictionary of users and unique questions
    uq_dict = {}
    # keys of dictionary (unique users)
    keys = []
    for k,v in recalls_new.groupby('user_id'):
        uq_dict[k] = set(v.question_id)
        keys.append(k)
    # for each combination of two users, determine how many questions they've both answered and update min_num as necessary
    for u1,u2 in itertools.combinations(keys, 2):
        n = len(uq_dict[u1].intersection(uq_dict[u2]))
        if n < min_num:
            min_num = n
    print "Minimum number of questions that any pair of users shares: " + str(min_num)
    
    return recalls_new
```

It appears that our data is too sparse to compare users from different institutions who have answered the same questions, since the minimum number of questions that any pair of users shares is 0 in all of the above cases. As with our previous method, we move up a level and look at topics that pairs of users share rather than questions that they share.

Again, we need to make sure that the minimum number of topics that any pair of users shares is not 0. The following function repeats what we have done earlier, except we replace the number of unique questions requirement with a number of unique topics requirement.

```python
"""
Given restraints on how many unique topics users need to have answered questions in to be considered a major user
and how many major users institutions need to have to be considered a major institution, print some useful
summary statistics. Also return the dataframe of all entries that remain after filtering by our requirements.

Parameters:
    recalls: the recalls dataframe to use
    nt: number of unique topics users need to have answered questions in to be considered a major user
    nu: number of major users institutions need to have to be considered a major institution

Returns recalls_new, a dataframe of all entries that remain after filtering by our requirements.
"""
def filter_recalls5(recalls, nt, nu):
    # filter recalls to only include users who have answered questions in at least nt topics
    grouped_u = recalls.groupby('user_id')
    recalls_new = grouped_u.filter(lambda x: x.topic_id.nunique() >= nt)
    
    print "Number of questions answered by major users: " + str(len(recalls_new))
    grouped_u = recalls_new.groupby('user_id')
    print "Number of major users: " + str(grouped_u.ngroups)
    
    # filter recalls_new to only include institutions where there are at least nu major users
    grouped_i = recalls_new.groupby('institution_id')
    recalls_new = grouped_i.filter(lambda x: x.user_id.nunique() >= nu)
    
    print "Number of questions answered by major users in major institutions: " + str(len(recalls_new))
    grouped_i = recalls_new.groupby('institution_id')
    print "Number of major institutions: " + str(grouped_i.ngroups)
    # number of users in major institutions
    n_users_final = len(set(recalls_new.user_id))
    print "Number of major users in major institutions: " + str(n_users_final)
    
    # find minimum number of topics that any pair of users has in common
    min_num = len(set(recalls_new.topic_id))
    # dictionary of users and unique topics
    ut_dict = {}
    # keys of dictionary (unique users)
    keys = []
    for k,v in recalls_new.groupby('user_id'):
        ut_dict[k] = set(v.topic_id)
        keys.append(k)
    # for each combination of two users, determine how many topics they share and update min_num as necessary
    for u1,u2 in itertools.combinations(keys, 2):
        n = len(ut_dict[u1].intersection(ut_dict[u2]))
        if n < min_num:
            min_num = n
    print "Minimum number of topics that any pair of users shares: " + str(min_num)
    
    return recalls_new
```

It appears that our data is too sparse to even compare users from different institutions who have answered questions in the same topics, since the minimum number of topics that any pair of users shares is 0 in all of the above cases. As with our previous method, we move up a level and look at subjects that pairs of users share rather than topics that they share.

Again, we need to make sure that the minimum number of subjects that any pair of users shares is not 0. The following function repeats what we have done earlier, except we replace the number of unique topics requirement with a number of unique subjects requirement.

```python
"""
Given restraints on how many unique subjects users need to have answered questions in to be considered a major user
and how many major users institutions need to have to be considered a major institution, print some useful
summary statistics. Also return the dataframe of all entries that remain after filtering by our requirements.

Parameters:
    recalls: the recalls dataframe to use
    ns: number of unique subjects users need to have answered questions in to be considered a major user
    nu: number of major users institutions need to have to be considered a major institution

Returns recalls_new, a dataframe of all entries that remain after filtering by our requirements.
"""
def filter_recalls6(recalls, ns, nu):
    # filter recalls to only include users who have answered questions in at least ns subjects
    grouped_u = recalls.groupby('user_id')
    recalls_new = grouped_u.filter(lambda x: x.subject.nunique() >= ns)
    
    print "Number of questions answered by major users: " + str(len(recalls_new))
    grouped_u = recalls_new.groupby('user_id')
    print "Number of major users: " + str(grouped_u.ngroups)
    
    # filter recalls_new to only include institutions where there are at least nu major users
    grouped_i = recalls_new.groupby('institution_id')
    recalls_new = grouped_i.filter(lambda x: x.user_id.nunique() >= nu)
    
    print "Number of questions answered by major users in major institutions: " + str(len(recalls_new))
    grouped_i = recalls_new.groupby('institution_id')
    print "Number of major institutions: " + str(grouped_i.ngroups)
    # number of users in major institutions
    n_users_final = len(set(recalls_new.user_id))
    print "Number of major users in major institutions: " + str(n_users_final)
    
    # find minimum number of subjects that any pair of users has in common
    min_num = len(set(recalls_new.subject))
    # dictionary of users and unique subjects
    us_dict = {}
    # keys of dictionary (unique users)
    keys = []
    for k,v in recalls_new.groupby('user_id'):
        us_dict[k] = set(v.subject)
        keys.append(k)
    # for each combination of two users, determine how many subjects they share and update min_num as necessary
    for u1,u2 in itertools.combinations(keys, 2):
        n = len(us_dict[u1].intersection(us_dict[u2]))
        if n < min_num:
            min_num = n
    print "Minimum number of subjects that any pair of users shares: " + str(min_num)
    
    return recalls_new
```

Finally, we have some results that we can work with, since in some of the cases the minimum number of subjects that any pair of users shares is nonzero. We'll use the data from the case where ns = 10 and nu = 3, since this gives us the most institutions and total users while still giving us a nonzero minimum number of subjects that any pair of users shares.

For each subject that a pair of users shares, we'll consider the average score of each user on questions in that subject. Since the difficulty of questions may be different, we'll divide users' scores on each question by the average score of that question. Since we're only considering users' initial scores on questions, the average will only be across the subset of data where the given user sees the question for the first time. Then take the average of the squared differences between the two users' averages in each subject, and use that as our distance. We average because different pairs of users may have different numbers of shared subjects, and we want a standardized metric. We also square the differences because we want to greater penalize cases where users are farther apart.

```python
# save the dataframe that we want
recalls_filtered = filter_recalls6(recalls, 10, 3)
```

The following is code that we ran to modify our dataframe so that we can easily filter for cases where it is the first time a user has seen a question. We stored the modified dataframe into a new csv for easier future reference.

```python
#modify our dataframe so that we can easily filter for cases where it is the first time a user has seen a question

#create a new column for number of times a question has been seen by the user seen
recalls['times_seen'] = [0] * len(recalls)

#group by user_id and question_id
grouped_uq = recalls.groupby(['user_id', 'question_id'])

for k, v in grouped_uq:
    #sort by time within each groupby
    v = v.sort('created_at')
    #keep track of how many times they have seen the question
    count = 0
    
    #iterate through all the times they have seen the question and update the times_seen column
    for i, r in v.iterrows():
        recalls.times_seen.ix[i] = count
        count += 1

# store into a csv for easier future reference
recalls.to_csv('recalls_for_graph.csv')
```

```python
# get updated recalls from csv
recalls_for_graph = pd.read_csv('recalls_for_graph.csv')
```

To create the graph, we will need to know the average score of each user within any subject. The following helper function calculates that for us.

```python
"""
A helper function that computes a given user's average score within a subject.
Average score is standardized by dividing each individual score on a question by the
average score on that question across all cases where a user first sees that question.

Parameters:
    recalls: the recalls dataframe to use
    user: the user id
    subject: the subject of interest

Returns the standardized average score for the given user in the given subject.
"""
def compute_avg_subj_score(recalls, user, subject):
    # consider only cases where it's the first time a user has seen a question
    recalls_new = recalls[recalls.times_seen == 0]
    # calculate user's average score across the subject
    total = 0.
    count = 0
    grouped_u = recalls_new.groupby('user_id')
    for k,v in grouped_u:
        if k == user:
            df_filtered = v[v.subject == subject]
            # create list of questions from questions column
            questions = []
            for q in df_filtered.question_id:
                questions.append(q)
            # create list of scores from score column
            scores = []
            for s in df_filtered.recall_score:
                scores.append(s)
            for i in range(0, len(questions)):
                # find average score of question
                avg = recalls_new[recalls_new.question_id == questions[i]].recall_score.mean()
                # add adjusted score of user to total, and increment count
                total += float(scores[i]) / avg
                count += 1
    return total / float(count)
```

Now we are ready to make the graph. We'll assign colors to each of the three institutions that are represented.

```python
# create graph
graph = nx.Graph()

# determine institutions
institutions = list(set(recalls_filtered.institution_id))
grouped_u = recalls_filtered.groupby('user_id')

# add all nodes
for k,v in grouped_u:
    graph.add_node(k)
    # determine institution of user and set color accordingly
    # we know there are 3 institutions from our earlier analysis
    if institutions[0] in set(v.institution_id):
        graph.node[k]['color'] = 'r'
    elif institutions[1] in set(v.institution_id):
        graph.node[k]['color'] = 'g'
    else:
        graph.node[k]['color'] = 'b'
        
# add edges
for u1, u2 in itertools.combinations(list(set(recalls_filtered.user_id)), 2):
    sum_squares = 0.
    count = 0
    for k,v in recalls_filtered.groupby('subject'):
        if (u1 in set(v.user_id)) and (u2 in set(v.user_id)):
            avg1 = compute_avg_subj_score(recalls_for_graph, u1, k)
            avg2 = compute_avg_subj_score(recalls_for_graph, u2, k)
            sum_squares += (avg1 - avg2)**2
            count += 1
    distance = sum_squares / float(count)
    # use the inverse of the distance as the weight of the edge
    graph.add_edge(u1, u2, weight = 1. / distance)
```

Finally, we draw the graph.  Since we were unable to gather the actual institutions' names or users' names due to Firecracker's privacy policy, we won't label our graph with user_id's or institution_id's, since the id's give no additional information to us.

```python
# this makes sure results are the same at each call
np.random.seed(1)  

color = [graph.node[user]['color'] for user in graph.nodes()]

#determine position of each node using a spring layout
pos = nx.spring_layout(graph, iterations=200)

#plot the edges
nx.draw_networkx_edges(graph, pos, alpha = .5)

#plot the nodes
nx.draw_networkx_nodes(graph, pos, node_color=color)

#coordinate information is meaningless here, so remove it
plt.xticks([])
plt.yticks([])
remove_border(left=False, bottom=False)
```

The spring layout tries to group nodes with large edge-weights near to each other, so it tries to organize the users into similarly-performing groups. From this visualization, it appears like the users may be grouped by institution: the green nodes are towards the bottom left of the graph, the blue nodes are towards the top right, and the red nodes are clumped around the center.

A way to verify if users tend to clump by institution is to examine the minimum spanning tree. When doing so, we'll try to minimize the distance between the nodes, which is just the inverse of the edge weights.

```python
# add distances, which we will use to calculate the minimum spanning tree
for (i,j) in graph.edges():
    graph[i][j]['distance'] = 1. / float(graph[i][j]['weight'])

# calculate minimum spanning tree
mst = nx.minimum_spanning_tree(graph, weight='distance')

# this makes sure results are the same at each call
np.random.seed(1)  

color = [graph.node[user]['color'] for user in graph.nodes()]

#determine position of each node using a spring layout
pos = nx.spring_layout(mst, iterations=200)

#plot the edges
nx.draw_networkx_edges(mst, pos, alpha = .5)

#plot the nodes
nx.draw_networkx_nodes(mst, pos, node_color=color)

#coordinate information is meaningless here, so remove it
plt.xticks([])
plt.yticks([])
remove_border(left=False, bottom=False)
```

From this minimum spanning tree we can also see some sort of similarities between users from the same institution. As in the earlier graph, users are generally closer to users from the same institution. This time there's the exception of the red user that is all the way at the other end of the tree from the other red users. In general, however, this tree shows that there are observable differences between institutions, which we expect to be true. It is reasonable to assume that different institutions have different focuses, and that two users within the same institution would score more similarly than two users from two different institutions.

We don't have nearly enough data to make any definitive conclusions, however. After making the data we work with less sparse and ensuring that we get distances between each node by only considering users that share subjects, we were left with only the 12 users above. In order to get a better picture, we need to examine more users, and to do so, we need more data. If we were able to get the full data set of all of Firecracker's users, we would likely be able to do a better analysis. But even if having the full dataset doesn't help too much, our process and results can still be meaningful to Firecracker because they demonstrate that there is likely a difference between institutions. Firecracker is still a relatively young service, so it may not have the wealth of data required to run more conclusive analyses on this matter. In the future, however, when Firecracker attracts more users and the existing users also have used the system more, we may be able to create a better graph that has more nodes from which we can draw more definitive conclusions.
